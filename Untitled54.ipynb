```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D  # noqa
plt.style.use("seaborn-v0_8")
```

```python
from google.colab import files
uploaded = files.upload()  # choose memoryforge_topics.csv from your PC

import io
import pandas as pd

df = pd.read_csv(io.BytesIO(uploaded["memoryforge_topics.csv"]))
df.head()
```

```python
room_names = df["group"].unique().tolist()
room_positions = {g: i for i, g in enumerate(room_names)}  # x index

coords = []
for _, row in df.iterrows():
    room_x = room_positions[row["group"]]
    # spread topics inside room
    y = np.random.uniform(-0.5, 0.5)
    # height encodes difficulty
    z = row["difficulty"] * 0.5
    coords.append((room_x, y, z))

df["x"], df["y"], df["z"] = zip(*coords)
df
```

```python
def plot_floorplan(df):
    fig, ax = plt.subplots(figsize=(10, 3))
    groups = df["group"].unique()
    for i, g in enumerate(groups):
        ax.add_patch(plt.Rectangle((i-0.4, -1), 0.8, 2, fill=False, edgecolor="black", linewidth=1.5))
        ax.text(i, 1.1, g, ha="center", va="bottom", fontsize=10, fontweight="bold")
        room_df = df[df["group"] == g]
        for _, r in room_df.iterrows():
            ax.scatter(i + r["y"], 0, s=40 + r["importance"]*10, alpha=0.8)
            ax.text(i + r["y"], 0.05, r["topic"], rotation=45, ha="left", va="bottom", fontsize=7)
    ax.set_ylim(-1.2, 1.5)
    ax.set_xlim(-0.5, len(groups)-0.5)
    ax.set_xticks([])
    ax.set_yticks([])
    ax.set_title("MemoryForge: Topdown Room Layout")
    plt.tight_layout()
    plt.show()

plot_floorplan(df)
```

```python
def plot_3d(df):
    fig = plt.figure(figsize=(8,6))
    ax = fig.add_subplot(111, projection="3d")

    for g in df["group"].unique():
        sub = df[df["group"] == g]
        ax.scatter(sub["x"], sub["y"], sub["z"], s=40 + sub["importance"]*10, label=g)

    for _, r in df.iterrows():
        ax.text(r["x"], r["y"], r["z"]+0.2, r["topic"], fontsize=7)

    ax.set_xlabel("Room index (forward/back)")
    ax.set_ylabel("Left / Right")
    ax.set_zlabel("Height (difficulty)")
    ax.set_title("MemoryForge: 3D Memory Palace Layout")
    ax.legend()
    plt.tight_layout()
    plt.show()

plot_3d(df)
```

```python
def describe_rooms(df):
    lines = []
    for g in df["group"].unique():
        room = df[df["group"] == g].sort_values(by="importance", ascending=False)
        lines.append(f"Room: {g}")
        core = room.iloc[0]["topic"]
        lines.append(f"  - Core wall concept: {core}")
        for _, r in room.iloc[1:].iterrows():
            pos = "upper wall" if r["difficulty"] >= room["difficulty"].median() else "side wall"
            lines.append(f"  - Place '{r['topic']}' on the {pos} (importance {r['importance']})")
        lines.append("")
    return "\n".join(lines)

print(describe_rooms(df))
```
